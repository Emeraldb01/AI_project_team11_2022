```python
import csv

file = open('keypoint.csv')

type(file)

csvreader = csv.reader(file)

rows = []
for row in csvreader:
        rows.append(row)
```

```python
classifier = []
data = []
for num in rows:
  tmp = []
  tmp0 = []

  tmp0.append(num[0])
  classifier.append(tmp0)
  
  tmp.append((float(num[10]) - float(num[4])) * 100) # tip of thumb - root of thumb
  tmp.append((float(num[18]) - float(num[12])) * 100) # tip of index - root of index
  data.append(tmp)

print(classifier)
print(data)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import Perceptron
import matplotlib.pyplot as plt

X = data
y = classifier

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)


clf = Perceptron(random_state=0)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

query = clf.predict([[-33.07086614173228, -44.09448818897637]])
print("Test run, the guess is : ")
print(query)
print("Overall accuracy:")
print(accuracy_score(y_test, y_pred))

# predict zero performance
X_test_0 = []
y_test_0 = []
for y in range(len(y_test)):
  if y_test[y][0] == '0':
    y_test_0.append(['0'])
    X_test_0.append(X_test[y])

y_pred_0 = clf.predict(X_test_0)
print("Accuracy predicting classifer 0: ")
print(accuracy_score(y_test_0, y_pred_0))

# predict one performance
X_test_1 = []
y_test_1 = []
for y in range(len(y_test)):
  if y_test[y][0] == '1':
    y_test_1.append(['1'])
    X_test_1.append(X_test[y])

y_pred_1 = clf.predict(X_test_1)
print("Accuracy predicting classifer 1: ")
print(accuracy_score(y_test_1, y_pred_1))

# predict 2 performance
X_test_2 = []
y_test_2 = []
for y in range(len(y_test)):
  if y_test[y][0] == '2':
    y_test_2.append(['2'])
    X_test_2.append(X_test[y])

y_pred_2 = clf.predict(X_test_2)
print("Accuracy predicting classifer 2: ")
print(accuracy_score(y_test_2, y_pred_2))

# predict 3 performance
X_test_3 = []
y_test_3 = []
for y in range(len(y_test)):
  if y_test[y][0] == '3':
    y_test_3.append(['3'])
    X_test_3.append(X_test[y])

y_pred_3 = clf.predict(X_test_3)
print("Accuracy predicting classifer 3: ")
print(accuracy_score(y_test_3, y_pred_3))

def Linear_classifier1(coord): # input all the x,y coordinate, coord[0] = x coordinate of 0, coord[1] = y coordinate of 0
  test = []

  x1 = (coord[9] - coord[3]) * 100
  x2 = (coord[17] - coord[11]) * 100

  test.append(x1)
  test.append(x2)

  prediction = clf.predict([test])
  return prediction
```
